```python
import numpy as np
import requests
from bs4 import BeautifulSoup
import pandas as pd

try:
  headers = {
    'User-Agent': 'Mozilla/5.0 (iPad; U; CPU OS 3_2_1 like Mac OS X; en-us) AppleWebKit/531.21.10 (KHTML, like Gecko) Mobile/7B405',
    'From': 'aw183052@terpmail.umd.edu'
  }
  
  r1 = requests.get("https://www.worldometers.info/world-population/us-population/", headers=headers)

except error as e:
  print(e)

root1 = BeautifulSoup(r1.content, "html")

#find and getting the table data for all 3 websites
webTable1 = root1.find('table', attrs = {"class": "table table-striped table-bordered table-hover table-condensed table-list"} ).prettify()

#creating a table for webtable1
columns = ['year', 'population', 'population percent change', 'yearly change', 'migrants', 'median age', 'fertility rate', 'density', 'urban pop percent', 'urban pop', 'country share of world pop', 'world pop', 'us global rank']
index = range(1, 19)

#takes the first dataframe from the list
dataTable1 = pd.read_html(webTable1)[0] #list of dataFrame

#set columns name to values in columns and set the index with range from 1 - 19
dataTable1.columns = columns
dataTable1.index = index 

dataTable1
```

```python
# Tiyding Data for datatable1

dataTable1 = dataTable1.drop(['migrants', 'median age', 'fertility rate', 'density', 'urban pop percent', 'urban pop', 'country share of world pop', 'world pop', 'us global rank'], axis = 1)

dataTable1
```

```python
# Tiyding Data for COVIDdata
dfChunk = pd.read_csv('https://covid19.who.int/WHO-COVID-19-global-data.csv', chunksize = 1000)
#print(data)
#data = data.drop(['Country_code', 'WHO_region', 'New_cases', 'New_deaths'], axis = 1)

chunk_list = []  # append each chunk df here 

# Each chunk is in df format
index = 0
for chunk in dfChunk:  


    for i in chunk.index:
        if index == 1000:
            index = 0
        if chunk["Country"][i] != 'United States of America':
            chunk = chunk.drop([i])
        else:
            index = index + 1
    # Once the data filtering is done, append the chunk to list
    chunk_list.append(chunk)
    
# concat the list into dataframe 
df_concat = pd.concat(chunk_list)
```

```python
numIndex = len(df_concat.index)

x = [*range(1,numIndex + 1)]
df_concat['index'] = x
df_concat = df_concat.set_index('index')
df_concat
```
